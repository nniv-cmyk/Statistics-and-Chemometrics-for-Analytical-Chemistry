```python
import numpy as np

# Assumed true value
true_value = 10.00

# Student results for each group
results = {
    'A': [10.08, 10.11, 10.09, 10.10, 10.12],
    'B': [9.88, 10.14, 10.02, 9.80, 10.21],
    'C': [10.19, 9.79, 9.69, 10.05, 9.78],
    'D': [10.04, 9.98, 10.02, 9.97, 10.04]
}

# Calculate mean, standard deviation, and bias for each group
for lab, values in results.items():
    mean = np.mean(values)
    std_dev = np.std(values, ddof=1)  # using Bessel's correction (ddof=1)
    bias = mean - true_value
    print(f"Group {lab} - Mean: {mean:.2f}, Std Dev: {std_dev:.3f}, Bias: {bias:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the mean and standard deviation
mu = 0  # Mean
sigma = 1  # Standard deviation

# Create a range of values from -4s to 4s around the mean
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Calculate the probability density function (PDF) for the normal distribution
y = norm.pdf(x, mu, sigma)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'k', linewidth=2)

# Fill the areas for 1s, 2s, and 3s
plt.fill_between(x, y, where=(mu - sigma <= x) & (x <= mu + sigma), color='blue', alpha=0.1, label='68% within ±1σ')
plt.fill_between(x, y, where=(mu - 2*sigma <= x) & (x <= mu + 2*sigma), color='green', alpha=0.1, label='95% within ±2σ')
plt.fill_between(x, y, where=(mu - 3*sigma <= x) & (x <= mu + 3*sigma), color='red', alpha=0.1, label='99.7% within ±3σ')

# Labeling the plot
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('Data Points')
plt.ylabel('Probability Density')
plt.legend(title='Legend')

# Show the plot
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm

# Mean and standard deviation
mu = 10.15
sigma = 0.02

# Range of interest
lower_bound = 10.12
upper_bound = 10.20

# Calculate the z-scores for the lower and upper bounds
z_lower = (lower_bound - mu) / sigma
z_upper = (upper_bound - mu) / sigma

# Calculate the probabilities for these z-scores using the CDF
p_lower = norm.cdf(z_lower)
p_upper = norm.cdf(z_upper)

# The proportion of values within the range
proportion_in_range = p_upper - p_lower

proportion_in_range
```
